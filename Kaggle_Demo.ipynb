# 🛒 Smart Retail B2B Procurement Guardrail
## Distributed Agent-to-Agent (A2A) Microservice Architecture

Welcome to the interactive Kaggle demonstration notebook for our **Business Agent Track** submission. 

This notebook is designed to replace a traditional local `client.py` script. It allows you to fire off a remote network request directly to our live **Render Cloud Deployments** and witness 4 isolated AI microservices successfully negotiate a corporate purchase order without human intervention.

```python
import httpx
import json
from IPython.display import display, JSON

# 1. Define our Entrypoint (The EDA Agent deployed on Render)
# This acts as the front-door to the entire A2A network.
EDA_URL = "https://neko-agent.onrender.com"
```

### ☁️ The Payload & Google Drive Integration
Instead of uploading a static CSV, we pass a **Google Drive URL** directly into the context payload. 
The EDA agent (running in the cloud) will dynamically simulate authentication via MCP, fetch the raw `.xlsx` file, convert it to `.csv`, calculate statistical variance, and pass the insight down the chain.

```python
initial_payload = {
    "session_id": "kaggle-demo-777",
    "history": [],
    "context": {
        "gdrive_url": "https://drive.google.com/file/d/1A2B3C4D5E6F7G8H9I0J/view?usp=sharing"
    }
}
```

### 🚀 Executing the A2A Pipeline
When we execute the cell below, the following distributed chain reaction occurs entirely over the public internet:
1. **EDA Agent:** Fetches GDrive file, finds a 15% sales anomaly.
2. **Sales Analyst:** Fuses the EDA math with unstructured store manager notes & Postgres database records.
3. **Strategist:** Queries wholesale supplier tools and drafts a Purchase Order.
4. **Guardrail:** Rejects any PO over budget, autonomously forcing the Strategist into a renegotiation retry loop.

*Note: Because this requires sequential Gemini API calls across 4 independent cloud servers, this cell may take 30-60 seconds to complete.*

```python
print(f"Initiating A2A Pipeline...\nSending payload to Front-Door Agent: {EDA_URL}/eda")

with httpx.Client(timeout=120.0) as client:
    try:
        response = client.post(f"{EDA_URL}/eda", json=initial_payload)
        result = response.json()
        print("\n✅ Pipeline Complete! The Guardrail has approved the final negotiation.\n")
    except Exception as e:
        print(f"\n❌ Pipeline failed or timed out: {e}")
        result = {"error": str(e)}
```

### 📊 View the Final Approved Purchase Order
The output below is the safe, budget-compliant result of 4 autonomous agents cross-checking each other. Notice how the Strategist negotiated the cost down to fit within the Guardrail's strict parameters.

```python
display(JSON(result))
```
